```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': {'$eq': 20}}

# Use count_documents to display the number of documents in the result
results = establishments.count_documents(query)
print("Number of establishments with a hygiene score of 20:", results)
print('----------------------------------------------------------------')

# Display the first document in the results using pprint
first_result = establishments.find_one(query)
pprint(first_result)
```

```python
# Convert the result to a Pandas DataFrame
hygiene_score_20 = pd.DataFrame(establishments.find(query))
hygiene_score_20
# Display the number of rows in the DataFrame
print(f"Number of rows in dataframe: {hygiene_score_20.shape[0]}")

# Display the first 10 rows of the DataFrame
hygiene_score_20.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {
    'AddressLine3': 'London',
    'RatingValue': {'$gte': 4}
}

# Use count_documents to display the number of documents in the result
results = establishments.count_documents(query)
print(f"Number of documents with London as Local Authority: ", results)
print('----------------------------------------------------------------')
# Display the first document in the results using pprint
pprint(establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
london_rating_40 = pd.DataFrame(list(establishments.find(query)))
london_rating_40
# Display the number of rows in the DataFrame
print(f"Number of rows in dataframe: {london_rating_40.shape[0]}")
print('----------------------------------------------------------------')

# Display the first 10 rows of the DataFrame
london_rating_40.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01

# Retrieve latitude and longitude values from the collection
latitude_cursor = establishments.find({}, {'geocode.latitude': 1})
longitude_cursor = establishments.find({}, {'geocode.longitude': 1})

# Extract latitude and longitude values from the cursors
latitude_values = [doc['geocode']['latitude'] for doc in latitude_cursor]
longitude_values = [doc['geocode']['longitude'] for doc in longitude_cursor]

latitude = latitude_values[0]
longitude = longitude_values[0]

query = {
    'geocode.latitude': {'$gte': latitude - degree_search, '$lte': latitude + degree_search},
    'geocode.longitude': {'$gte': longitude - degree_search, '$lte': longitude + degree_search},
    'RatingValue': 5
}
sort = [('scores.Hygiene', 1)]

# Print the results
results = list(establishments.find(query).sort(sort))
pprint(results)
```

```python
# Convert result to Pandas DataFrame
lat_lng_df = pd.DataFrame(results)
lat_lng_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match_query = {'$match': {'count': {'$eq':5}}}
# 2. Groups the matches by Local Authority
group_query = {'$group':{'_id': {'AddressLine3': '$AddressLine3'},
                        'count': {'$sum':1}}}
# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': {'_id.AddressLine3':1, 'count':-1}}

pipeline = [group_query,match_query,sort_values]
# Print the number of documents in the result
results = list(establishments.aggregate(pipeline))

count = sum(1 for result in results)
print(f"Number of documents in the result: ", count)
print('-------------------------------------------------')

# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
local_authority_5 = pd.DataFrame(results)
local_authority_5
# Display the number of rows in the DataFrame
print(f"Number of rows in dataframe: {local_authority_5.shape[0]}")
print('----------------------------------------------------------------')
# Display the first 10 rows of the DataFrame
local_authority_5.head(10)
```
